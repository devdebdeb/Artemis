```python
estatisticas_full['win'] = estatisticas_full['win'].replace({True: 1, False: 0})
```

```python
estatisticas_full = estatisticas_full.drop(columns=['id'])
estatisticas_full = estatisticas_full.drop(columns=['getTakedownsInAllLanesEarlyJungleAsLaner'])
estatisticas_full = estatisticas_full.drop(columns=['totalEnemyJungleMinionsKilledwin'])
```

```python
estatisticas_full_semff = estatisticas_full[estatisticas_full['gameLength'] >= 899]
estatisticas_full_semff = estatisticas_full[estatisticas_full['jungleCsBefore10Minutes'] >= 20]
```

```python
df_para_regressão = {key: value for key, value in estatisticas_full_semff.groupby('Main')}
```

```python
# Definindo o caminho da pasta onde os arquivos serão salvos
output_dir = '/content/dataframes'

# Criando a pasta se ela não existir
os.makedirs(output_dir, exist_ok=True)
```

```python
for key, value in df_para_regressão.items():
    filename = os.path.join(output_dir, f'dataframe_{key}.xlsx')
    value.to_excel(filename, index=False)
    print(f"DataFrame para main = {key} salvo como {filename}")
```
